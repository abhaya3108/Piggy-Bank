```python
#Write a program in Python with one class called Cipher
import random
class Cipher():
    key = random.randint(1,51)
        
    def __init__(self, text):
        self.text = text
        
    def encrypt(self, text): 
        result = ""

        # traverse text 
        for i in range(len(text)): 
            char = text[i] 
            
            # Encrypt uppercase characters 
            if (char.isupper()): 
                result += chr((ord(char) + Cipher.key - 65) % 26 + 65) 

            # Encrypt lowercase characters 
            else: 
                result += chr((ord(char) + Cipher.key - 97) % 26 + 97) 

        return result 

    def decrypt(self, text): 
        result = "" 

        # traverse text 
        for i in range(len(text)): 
            char = text[i] 
            
            # Encrypt uppercase characters 
            if (char.isupper()): 
                result += chr((ord(char) - Cipher.key - 65) % 26 + 65) 

            # Encrypt lowercase characters 
            else: 
                result += chr((ord(char) - Cipher.key - 97) % 26 + 97) 

        return result 

c = Cipher('Cipher')
print(c.encrypt('Cipher'))
print(c.decrypt(c.encrypt('Cipher')))
```
